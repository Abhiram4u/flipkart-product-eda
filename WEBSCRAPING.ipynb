```python
Brand = []
Capacity = []
CoolingTechnology = []
DesignType = []
EnergyEfficiencyRating = []
CompressorType = []
ProductWarranty = []
CompressorWarranty = []
Rating = []
Price = []

for x in range(1,19):
    url = f"https://www.flipkart.com/search?q=fridges&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DLG&p%5B%5D=facets.brand%255B%255D%3DWhirlpool&p%5B%5D=facets.brand%255B%255D%3DHaier&p%5B%5D=facets.brand%255B%255D%3DGodrej&p%5B%5D=facets.brand%255B%255D%3DBlue%2BStar&p%5B%5D=facets.rating%255B%255D%3D3%25E2%2598%2585%2B%2526%2Babove&page={x}"
    page = requests.get(url,headers = request_header)

    soup = BeautifulSoup(page.text)

    for i in soup.find_all("div",class_ = "KzDlHZ"):
        text = i.text

        regex = re.findall("^[A-Za-z]+",text)
    
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)

        regex = re.findall("\s([0-9]+)\sL",text)
    
        if regex:
            Capacity.append(regex[0])
        else:
            Capacity.append(np.nan)
        
        regex = re.findall("Frost Free|Direct Cool",text)
    
        if regex:
            CoolingTechnology.append(regex[0])
        else:
            CoolingTechnology.append(np.nan)

        regex = re.findall("Double Door|Single Door|Triple Door|Side by Side",text)
    
        if regex:
            DesignType.append(regex[0])
        else:
            DesignType.append(np.nan)

        regex = re.findall("([0-9]) Star",text)
    
        if regex:
            EnergyEfficiencyRating.append(regex[0])
        else:
            EnergyEfficiencyRating.append(np.nan)
    for i in soup.find_all("div",class_="tUxRFH"):
        text = i.text
        regex = re.findall("Inverter|Digital Inverter|Smart Inverter|Reciprocatory|Reciprocating|Normal Compressor",text)
        if regex:
            CompressorType.append(regex[0])
        else:
            CompressorType.append(np.nan)

        regex = re.findall("r([0-9])\s",text)
        if regex:
            ProductWarranty.append(regex[0])
        else:
            ProductWarranty.append(np.nan)

        regex = re.findall("\s([0-9][0-9]|[0-9])\s[Y|y]ears",text)
        if regex:
            CompressorWarranty.append(regex[0])
        else:
            CompressorWarranty.append(np.nan)
    for i in soup.find_all("div",class_="XQDdHH"):
        Rating.append(i.text)

    for i in soup.find_all("div",class_="tUxRFH"):
        Price.append(i.find("div",class_="Nx9bqj _4b5DiR").text)

  

    
```

```python
df = pd.DataFrame({"Brand":Brand,"EnergyEfficiencyRating":EnergyEfficiencyRating,"DesignType": DesignType,"CoolingTechnology":CoolingTechnology\
              ,"Capacity(in Litres)":Capacity,"CompressorType":CompressorType,"Rating":Rating,"ProductWarranty(in years)":ProductWarranty\
              ,"CompressorWarranty(in years)":CompressorWarranty,"Price":Price})
```
